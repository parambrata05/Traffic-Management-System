```python
import pandas as pd
import matplotlib.pyplot as plt

# (optional but recommended for cleaner plots)
plt.style.use('default')

df = pd.read_csv("../data/processed/traffic_data_cleaned.csv")
df.head()
```

```python
df["traffic_level"].value_counts().plot(kind="bar")

plt.title("Traffic Level Distribution", fontsize=14)
plt.xlabel("Traffic Level")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
